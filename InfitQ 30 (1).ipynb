```python
#35
#lex_auth_0127136447430656001216

def convert_to_roman(num):
    #Start writing your code here
    n=str(num)
    d1={1:'I',2:'II',3:'III',4:'IV',5:'V',6:'VI',7:'VII',8:'VIII',9:'IX',10:'X',50:'L',100:'C',500:'D',1000:'M'}
    l=''
    def TD(n):
        if len(n)==1:
            return d1[int(n)]
        elif len(n)==2 and int(n)<=49:
            if n[1]=='0' and n!="40":
                l='X'*int(n[0])
                return l
            elif n=="40":
                return 'XL'
            elif int(n)>40 and int(n)<50:
                l="XL"+d1[int(n[1])]
                return l
            else:
                l='X'*int(n[0])+d1[int(n[1])]
                return l 
        elif len(n)==2 and int(n)>49:
            if n[1]=='0' and n!="90":
                l='L'+'X'*int(int(n[0])-5)
                return l
            elif n=="50":
                return 'l'
            elif n=="90":
                return 'XC'
            elif int(n)>90 and int(n)<100:
                l="XC"+d1[int(n[1])]
                return l
            else:
                l='L'+'X'*int(int(n[0])-5)+d1[int(n[1])]
                return l 
    def THD(n):
        if len(n)==3 and n[1]==n[2]=='0':
            l='C'*int(n[0])
            return l
        elif len(n)==3 and n[1]=='0':
            l='C'*int(n[0])+d1[int(n[2])]
            return l
        else:
            l='C'*int(n[0])+TD(n[1:])
            return l
    def THD1(n):
        if len(n)==3 and n[1]==n[2]=='0':
            if n=='400':
                return 'XD'
            elif n=="500":
                return "D"
            elif n=="900":
                return "CM"
            elif int(n)>500 and int(n)<900:
                l='D'+"C"*(int(n[0])-5)
                return l
        elif len(n)==3 and n[1]=='0':
            if int(n)<500:
                l='XD'+d1[int(n[2])]
                return l
            elif int(n)>500 and int(n)<900:
                l='D'+"C"*(int(n[0])-5)+d1[int(n[2])]
                return l
            elif int(n)>900 and int(n)<1000:
                l='CM'+TD(n[1:])
                return l
        else:
            if int(n)<500:
                l='CD'+TD(n[1:])
                return l
            elif int(n)>500 and int(n)<900:
                l='D'+"C"*(int(n[0])-5)+TD(n[1:])
                return l
            elif int(n)>900 and int(n)<1000:
                l='CM'+TD(n[1:])
                return l
    def FD(n):
        if n=='1000':
            return 'M'
        elif len(n)==4 and n[1]==n[2]==n[3]=='0':
            l='M'*int(n[0])
            return l
        elif len(n)==4 and n[2]==n[3]=='0':
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l
        elif len(n)==4 and n[3]=='0':
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l
        elif len(n)==4 and n[1]==n[2]=='0':
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l
        elif len(n)==4 and n[1]==n[3]=='0':
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l
        elif len(n)==4 and n[2]=='0':
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l
        elif len(n)==4 and n[1]=='0':
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l
        else:
            if int(n[1:])<400:
                l='M'*int(n[0])+THD(n[1:])
                return l
            elif int(n[1:])>=400:
                l='M'*int(n[0])+THD1(n[1:])
                return l       
            
    if len(n)<=2:
        return TD(n)
    elif len(n)==3 and int(n)<400:
        return THD(n)
    elif len(n)==3 and int(n)>=400:
        return THD1(n)
    elif len(n)==4 and int(n)>=1000:
        return FD(n)
        
        
        
        
    
    


for num in range(1,5000):    
    print(num," : ",convert_to_roman(num))
```

```python
def list_rotate(uniquecode_list):
    rotated_list=[]
    a=''
    count=0
    #Write your code here
    for i in uniquecode_list:
        for j in i:
            if j.isalpha():
                a+=j
                #print(a)
                count+=1
                #print(count)
        if count>=2:
            a+=i[-2:]+i[-4:-2]
            rotated_list.append(a)
            a=''
            count=0
        elif count==1:
            a+=i[-3:]+i[-4:-3]
            rotated_list.append(a)
            a=''
            count=0    
    return rotated_list

#You may modify the below code for testing
uniquecode_list=["AZ01-1234","R137-8714"]
rotated_list = list_rotate(uniquecode_list)
print(rotated_list)
```
